```python
import pandas as pd
%load_ext autoreload
%autoreload 2
dfgat = pd.DataFrame()
for chunk in pd.read_csv('/Users/maulik/Dropbox (HMS)/BrCaLines_profiling/ProfilingPaper/cell_cycle_data/merged/batches1to11_gating.results.csv', chunksize=500):
    df = chunk
    df['D'] = df['cell_count__dead']#.div(df['cell_count__total'])
    df[['G1', 'S', 'G2', 'M']] = df[['G1', 'S', 'G2', 'M']].multiply(df['cell_count'], axis=0)
    dfgat = dfgat.append(df)
dfgat = dfgat.replace('72', 72)
dfgat = dfgat.replace('time0_ctrl', 0)
dfgat.head()
```

```python
cell_line = ['MCF-12A', 
             '184B5',
             'ESVA-T',
             'MDA-MB-175-VII',
             'SUM185PE']

agent = ['Alpelisib/BYL719', 
         'Buparlisib/NVP-BKM120',
         'Pictilisib/GDC0941',
         'Taselisib/GDC0032',
         'Abemaciclib/LY2835219',
         'Palbociclib/PD0332991', 
         'LEE011/Ribociclib', 
         'Flavopiridol']
```

```python
cell = cell_line[0]
drug = agent[0]
dfctrl = dfgat[(dfgat.cell_line==cell) & (dfgat.agent=='DMSO') & (dfgat.timepoint==0)][['timepoint', 'concentration', 'well', 'G1', 'S', 'G2', 'M', 'D']]
dftreat = dfgat[(dfgat.cell_line==cell) & (dfgat.agent==drug)][['timepoint', 'concentration', 'well', 'G1', 'S', 'G2', 'M', 'D']]
```

```python
from random import randrange, uniform
import numpy as np
from scipy.optimize import minimize
from cell_cycle import obj_func, ode_soln

conc = dftreat.concentration.unique().tolist()
for conc in [conc[0]]:
    grpconc = dftreat.groupby('concentration', as_index=False)
    dfdata = pd.DataFrame()
    wells = []
    for i in range(grpconc.get_group(conc).shape[0]):
        dfdata = dfdata.append(dfctrl[dfctrl.well==dfctrl['well'].tolist()[randrange(len(dfctrl['well'].tolist()))]])
    dfdata = dfdata.append(grpconc.get_group(conc))
    dfdata = dfdata.reset_index().drop(columns=['index', 'concentration', 'well'])
    dfdata = dfdata.groupby('timepoint', as_index=False).mean()
    print dfdata
    #cell cycle at t=0
    x0 = dfdata[dfdata.timepoint==0][['G1', 'S', 'G2', 'M', 'D']].values.tolist()[0]
    #inital guess of parameters
    for i in range(10):
        logk0 = [np.log10(uniform(0.05,0.15)), np.log10(uniform(0.05,0.15)), np.log10(uniform(0.05,0.15)), np.log10(uniform(0.05,0.15)), np.log10(uniform(0.0005,0.001))]
        fit_rates = minimize(obj_func, logk0, args=(dfdata))
        print (fit_rates.x.tolist(), fit_rates.fun)
        print (ode_soln(x0, dfdata.timepoint.tolist(), fit_rates.x.tolist())[-1])
```
